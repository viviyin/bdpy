```python
import matplotlib.pyplot as plt

x = range(10)
plt.plot(x,x)
```

```python
import numpy as np
ax = np.array(x)
ay = ax**2
# r,g,b,c,m,y,k
# -, ., o, *, s, ^
# c^--
plt.xlabel('cost')
plt.ylabel('performance')
plt.axis([-10,10,-20,100])
plt.plot(ax,ay, 'c^--')
```
