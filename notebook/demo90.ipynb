```python
import pandas as pd
import numpy as np
np.random.seed(20200717)
df1 = pd.DataFrame(np.random.randn(6,7),
                  index=list(range(0,12,2)),
                  columns=list(range(0,7,1)))
df1
```

```python
func1 = lambda x: x.max()-x.min()
df1.apply(func1)
```

```python
df1.apply(func1, axis=1)
```

```python
df1.apply(func1, axis='columns')
```

```python
func2 = lambda x: pd.Series([x.min(), x.max()], index=['min','max'])
df1.apply(func2)
```

```python
df1.apply(func2, axis=1)
```

```python
df1.index = df1.iloc[:,1]
df1
```

```python
df1.sort_index()
```

```python
df1.columns = df1.iloc[2,:]
df1.columns
df1
```

```python
df1.sort_index(axis=1)
```
